```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga de archivo
df = pd.read_csv("listings_ROME.csv", engine='python', quotechar='"')
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
## Identificar valores nulos
pd.set_option('display.max_rows', None)
display(df.isnull().sum())
```

```python
## Copia
df_ROMA = df
df_ROMA.head(5)
```

**VALORES NULOS**

```python
#Establecer un nombre específico "Unknown"/"NONE" a todas los valores nulos dentro de las columnas 'description', 'neighborhood_overview', entre otras
df_ROMA["description"] = df_ROMA["description"].fillna("NONE")
df_ROMA["neighborhood_overview"] = df_ROMA["neighborhood_overview"].fillna("NONE")
df_ROMA["host_name"] = df_ROMA["host_name"].fillna("UNKOWN")
df_ROMA["host_since"] = df_ROMA["host_since"].fillna("00/00/00")
df_ROMA["host_location"] = df_ROMA["host_location"].fillna("Rome, Italy")
df_ROMA["host_about"] = df_ROMA["host_about"].fillna("NONE")

display(df_ROMA.isnull().sum())
```

```python
## LLenamos los NAs de la columna de 'host_response_time' con el string más repetido; "within an hour"
valor_mas_repetido_response = df_ROMA["host_response_time"].mode()[0]
df_ROMA["host_response_time"] = df_ROMA["host_response_time"].fillna(valor_mas_repetido_response)

#Eliminar un signo de columna 'host_response_rate' en dataframe df (copia)
if df_ROMA['host_response_rate'].dtype == 'object':
    df_ROMA['host_response_rate']=df_ROMA['host_response_rate'].str.replace('%', '')

#Eliminar un signo de columna 'host_response_rate' en dataframe df (original)
if df['host_response_rate'].dtype == 'object':
    df['host_response_rate']=df['host_response_rate'].str.replace('%', '')

#Conversión de tipo de dato de columna de tipo Object a Float
df_ROMA['host_response_rate']= df_ROMA['host_response_rate'].astype(float)
df['host_response_rate']= df['host_response_rate'].astype(float)

#Sustituir valores nulos con promedio o media en columna 'host_response_rate'
df_ROMA["host_response_rate"]=df_ROMA["host_response_rate"].fillna(round(df["host_response_rate"].mean(),1))
display(df_ROMA.isnull().sum())

display(df_ROMA.isnull().sum())
```

```python
#Eliminar un signo de columna 'host_acceptance_rate' en dataframe df (copia)
if df_ROMA['host_acceptance_rate'].dtype == 'object':
    df_ROMA['host_acceptance_rate']=df_ROMA['host_acceptance_rate'].str.replace('%', '')

#Eliminar un signo de columna 'host_acceptance_rate' en dataframe df (original)
if df['host_acceptance_rate'].dtype == 'object':
    df['host_acceptance_rate']=df['host_acceptance_rate'].str.replace('%', '')

#Conversión de tipo de dato de columna de tipo Object a Float
df_ROMA['host_acceptance_rate']= df_ROMA['host_acceptance_rate'].astype(float)
df['host_acceptance_rate']= df['host_acceptance_rate'].astype(float)

#Sustituir valores nulos con promedio o media en columna 'host_acceptance_rate'
df_ROMA["host_acceptance_rate"]=df_ROMA["host_acceptance_rate"].fillna(round(df["host_acceptance_rate"].mean(),1))
display(df_ROMA.isnull().sum())
```

```python
#Agregar nueva categoría 'NONE' a celdas vacias dentro de columna 'host_is_superhost' y 'host_identity_verified'
df_ROMA["host_is_superhost"] = df_ROMA["host_is_superhost"].fillna("NONE")
df_ROMA["host_thumbnail_url"] = df_ROMA["host_thumbnail_url"].fillna("UNKNOWN")
df_ROMA["host_picture_url"] = df_ROMA["host_picture_url"].fillna("UNKNOWN")
df_ROMA["host_neighbourhood"] = df_ROMA["host_neighbourhood"].fillna("UNKNOWN")
df_ROMA["host_verifications"] = df_ROMA["host_verifications"].fillna("NONE")
df_ROMA["host_has_profile_pic"] = df_ROMA["host_has_profile_pic"].fillna("NONE")
df_ROMA["host_identity_verified"] = df_ROMA["host_identity_verified"].fillna("NONE")

display(df_ROMA.isnull().sum())
```

```python
# Calculate the median for 'host_listings_count' and 'host_total_listings_count'
median_host_listings_count = df_ROMA['host_listings_count'].median()
median_host_total_listings_count = df_ROMA['host_total_listings_count'].median()

# Fill missing values with the calculated medians
df_ROMA['host_listings_count'] = df_ROMA['host_listings_count'].fillna(median_host_listings_count)
df_ROMA['host_total_listings_count'] = df_ROMA['host_total_listings_count'].fillna(median_host_total_listings_count)

## LLenamos los NAs de la columna de 'neighbourhood' con el vecindario más repetido; "Rome, Lazio, Italy"
valor_mas_repetido_response = df_ROMA["neighbourhood"].mode()[0]
print("El valor más repetido es:", valor_mas_repetido_response)

df_ROMA["neighbourhood"] = df_ROMA["neighbourhood"].fillna(valor_mas_repetido_response)

# Display the null value counts to verify
display(df_ROMA.isnull().sum())
```

```python
df_ROMA = df_ROMA.drop('neighbourhood_group_cleansed', axis=1)
display(df_ROMA.isnull().sum())
```

```python
#Establecer con mediana NAs de la columna 'bedrooms', 'bathrooms' y 'beds'
df_ROMA["bathrooms"]=df_ROMA["bathrooms"].fillna(round(df["bathrooms"].median(),1))
df_ROMA["bedrooms"]=df_ROMA["bedrooms"].fillna(round(df["bedrooms"].median(),1))
df_ROMA["beds"]=df_ROMA["beds"].fillna(round(df["beds"].median(),1))

#Agregar nueva categoría 'NONE' a celdas vacias dentro de columna 'bathrooms_text '
df_ROMA["bathrooms_text"] = df_ROMA["bathrooms_text"].fillna("NONE")



display(df_ROMA.isnull().sum())
```

```python
# Convertir la columna 'price' a numérica (eliminando símbolos de $ y comas)
df_ROMA["price"] = df["price"].replace('[\$,]', '', regex=True).astype(float)

# Calcular el promedio de precio por tipo de cuarto
promedios = df_ROMA.groupby("room_type")["price"].mean()
promedios

# Rellenar los NaN en 'price' con el promedio del room_type correspondiente
df_ROMA["price"] = df_ROMA.apply(
    lambda row: promedios[row["room_type"]] if pd.isna(row["price"]) else row["price"], axis=1)
```

```python
columns_to_fill = [
    'minimum_nights',
    'maximum_nights',
    'minimum_minimum_nights',
    'maximum_minimum_nights',
    'minimum_maximum_nights',
    'maximum_maximum_nights',
    'minimum_nights_avg_ntm',
    'maximum_nights_avg_ntm'
]

for col in columns_to_fill:
    median_value = df_ROMA[col].median()
    df_ROMA[col] = df_ROMA[col].fillna(median_value)

display(df_ROMA.isnull().sum())
```

```python
# Calculate the mode for 'has_availability'
mode_has_availability = df_ROMA['has_availability'].mode()[0]

# Fill missing values with the calculated mode
df_ROMA['has_availability'] = df_ROMA['has_availability'].fillna(mode_has_availability)

display(df_ROMA.isnull().sum())
```

```python
# Calculate the median for 'estimated_revenue_l365d'
median_estimated_revenue = df_ROMA['estimated_revenue_l365d'].median()

# Fill missing values with the calculated median
df_ROMA['estimated_revenue_l365d'] = df_ROMA['estimated_revenue_l365d'].fillna(median_estimated_revenue)

# Establecer un código de fecha para first y last_review
df_ROMA["first_review"] = df_ROMA["first_review"].fillna("00/00/00")
df_ROMA["last_review"] = df_ROMA["last_review"].fillna("00/00/00")

#Establecer con 0 todos los values que no cuentan con un review dentro este tipo de variables el dataframe
df_ROMA['reviews_per_month'] = df_ROMA['reviews_per_month'].fillna(0)
df_ROMA['review_scores_rating'] = df_ROMA['review_scores_rating'].fillna(0)
df_ROMA['review_scores_accuracy'] = df_ROMA['review_scores_accuracy'].fillna(0)
df_ROMA['review_scores_cleanliness'] = df_ROMA['review_scores_cleanliness'].fillna(0)
df_ROMA['review_scores_checkin'] = df_ROMA['review_scores_checkin'].fillna(0)
df_ROMA['review_scores_communication'] = df_ROMA['review_scores_communication'].fillna(0)
df_ROMA['review_scores_location'] = df_ROMA['review_scores_location'].fillna(0)
df_ROMA['review_scores_value'] = df_ROMA['review_scores_value'].fillna(0)

#Establecer una valor "UNKNOWN" a todas los valores nulos dentro de la columna 'license'
df_ROMA["license"] = df_ROMA["license"].fillna("UNKNOWN")

display(df_ROMA.isnull().sum())
```

**PROCEDIMIENTO "IQR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
# Display data types of columns in df_quantitative to identify non-numeric columns
print("Data types in df_quantitative:")
print(df_quantitative.dtypes)

# Select only numeric columns for IQR calculation and outlier treatment
numeric_df_quantitative = df_quantitative.select_dtypes(include=np.number).copy()

# Calculate Q1, Q3, and IQR for each numeric column
Q1 = numeric_df_quantitative.quantile(0.25)
Q3 = numeric_df_quantitative.quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Replace outliers with the median for each column
for col in numeric_df_quantitative.columns:
    median_value = numeric_df_quantitative[col].median()
    numeric_df_quantitative[col] = np.where(
        (numeric_df_quantitative[col] < lower_bound[col]) | (numeric_df_quantitative[col] > upper_bound[col]),
        median_value,
        numeric_df_quantitative[col]
    )

df_quantitative_filled_outliers = numeric_df_quantitative

display(df_quantitative_filled_outliers.head())
display(df_quantitative_filled_outliers.shape)
```
