```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga de archivo
df = pd.read_csv("listings_ROME.csv", engine='python', quotechar='"', on_bad_lines='skip')
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
## Identificar valores nulos
pd.set_option('display.max_rows', None)
display(df.isnull().sum())
```

```python
## Copia
df_ROMA = df
df_ROMA.head(5)
```

**VALORES NULOS**

```python
#Establecer un nombre específico "Unknown"/"NONE" a todas los valores nulos dentro de las columnas 'description', 'neighborhood_overview', entre otras
df_ROMA["description"] = df_ROMA["description"].fillna("NONE")
df_ROMA["neighborhood_overview"] = df_ROMA["neighborhood_overview"].fillna("NONE")
df_ROMA["host_name"] = df_ROMA["host_name"].fillna("UNKOWN")
df_ROMA["host_since"] = df_ROMA["host_since"].fillna("00/00/00")
df_ROMA["host_location"] = df_ROMA["host_location"].fillna("Rome, Italy")
df_ROMA["host_about"] = df_ROMA["host_about"].fillna("NONE")

display(df_ROMA.isnull().sum())
```

```python
## LLenamos los NAs de la columna de 'host_response_time' con el string más repetido; "within an hour"
valor_mas_repetido_response = df_ROMA["host_response_time"].mode()[0]
df_ROMA["host_response_time"] = df_ROMA["host_response_time"].fillna(valor_mas_repetido_response)

#Eliminar un signo de columna 'host_response_rate' en dataframe df (copia)
if df_ROMA['host_response_rate'].dtype == 'object':
    df_ROMA['host_response_rate']=df_ROMA['host_response_rate'].str.replace('%', '')

#Eliminar un signo de columna 'host_response_rate' en dataframe df (original)
if df['host_response_rate'].dtype == 'object':
    df['host_response_rate']=df['host_response_rate'].str.replace('%', '')

#Conversión de tipo de dato de columna de tipo Object a Float
df_ROMA['host_response_rate']= df_ROMA['host_response_rate'].astype(float)
df['host_response_rate']= df['host_response_rate'].astype(float)

#Sustituir valores nulos con promedio o media en columna 'host_response_rate'
df_ROMA["host_response_rate"]=df_ROMA["host_response_rate"].fillna(round(df["host_response_rate"].mean(),1))
display(df_ROMA.isnull().sum())

display(df_ROMA.isnull().sum())
```

```python
#Eliminar un signo de columna 'host_acceptance_rate' en dataframe df (copia)
if df_ROMA['host_acceptance_rate'].dtype == 'object':
    df_ROMA['host_acceptance_rate']=df_ROMA['host_acceptance_rate'].str.replace('%', '')

#Eliminar un signo de columna 'host_acceptance_rate' en dataframe df (original)
if df['host_acceptance_rate'].dtype == 'object':
    df['host_acceptance_rate']=df['host_acceptance_rate'].str.replace('%', '')

#Conversión de tipo de dato de columna de tipo Object a Float
df_ROMA['host_acceptance_rate']= df_ROMA['host_acceptance_rate'].astype(float)
df['host_acceptance_rate']= df['host_acceptance_rate'].astype(float)

#Sustituir valores nulos con promedio o media en columna 'host_acceptance_rate'
df_ROMA["host_acceptance_rate"]=df_ROMA["host_acceptance_rate"].fillna(round(df["host_acceptance_rate"].mean(),1))
display(df_ROMA.isnull().sum())
```

```python
#Agregar nueva categoría 'NONE' a celdas vacias dentro de columna 'host_is_superhost' y 'host_identity_verified'
df_ROMA["host_is_superhost"] = df_ROMA["host_is_superhost"].fillna("000")
df_ROMA["host_thumbnail_url"] = df_ROMA["host_thumbnail_url"].fillna("UNKNOWN")
df_ROMA["host_picture_url"] = df_ROMA["host_picture_url"].fillna("UNKNOWN")
df_ROMA["host_neighbourhood"] = df_ROMA["host_neighbourhood"].fillna("UNKNOWN")
df_ROMA["host_verifications"] = df_ROMA["host_verifications"].fillna("NONE")
df_ROMA["host_has_profile_pic"] = df_ROMA["host_has_profile_pic"].fillna("NONE")
df_ROMA["host_identity_verified"] = df_ROMA["host_identity_verified"].fillna("NONE")

display(df_ROMA.isnull().sum())
```

```python
# Calculate the median for 'host_listings_count' and 'host_total_listings_count'
median_host_listings_count = df_ROMA['host_listings_count'].median()
median_host_total_listings_count = df_ROMA['host_total_listings_count'].median()

# Fill missing values with the calculated medians
df_ROMA['host_listings_count'] = df_ROMA['host_listings_count'].fillna(median_host_listings_count)
df_ROMA['host_total_listings_count'] = df_ROMA['host_total_listings_count'].fillna(median_host_total_listings_count)

## LLenamos los NAs de la columna de 'neighbourhood' con el vecindario más repetido; "Rome, Lazio, Italy"
valor_mas_repetido_response = df_ROMA["neighbourhood"].mode()[0]
print("El valor más repetido es:", valor_mas_repetido_response)

df_ROMA["neighbourhood"] = df_ROMA["neighbourhood"].fillna(valor_mas_repetido_response)

# Display the null value counts to verify
display(df_ROMA.isnull().sum())
```

```python
df_ROMA = df_ROMA.drop('neighbourhood_group_cleansed', axis=1)
display(df_ROMA.isnull().sum())
```

```python
#Establecer con mediana NAs de la columna 'bedrooms', 'bathrooms' y 'beds'
df_ROMA["bathrooms"]=df_ROMA["bathrooms"].fillna(round(df["bathrooms"].median(),1))
df_ROMA["bedrooms"]=df_ROMA["bedrooms"].fillna(round(df["bedrooms"].median(),1))
df_ROMA["beds"]=df_ROMA["beds"].fillna(round(df["beds"].median(),1))

#Agregar nueva categoría 'NONE' a celdas vacias dentro de columna 'bathrooms_text '
df_ROMA["bathrooms_text"] = df_ROMA["bathrooms_text"].fillna("NONE")



display(df_ROMA.isnull().sum())
```

```python
# Convertir la columna 'price' a numérica (eliminando símbolos de $ y comas)
df_ROMA["price"] = df["price"].replace('[\$,]', '', regex=True).astype(float)

# Calcular el promedio de precio por tipo de cuarto
promedios = df_ROMA.groupby("room_type")["price"].mean()
promedios

# Rellenar los NaN en 'price' con el promedio del room_type correspondiente
df_ROMA["price"] = df_ROMA.apply(
    lambda row: promedios[row["room_type"]] if pd.isna(row["price"]) else row["price"], axis=1)
```

```python
columns_to_fill = [
    'minimum_nights',
    'maximum_nights',
    'minimum_minimum_nights',
    'maximum_minimum_nights',
    'minimum_maximum_nights',
    'maximum_maximum_nights',
    'minimum_nights_avg_ntm',
    'maximum_nights_avg_ntm'
]

for col in columns_to_fill:
    median_value = df_ROMA[col].median()
    df_ROMA[col] = df_ROMA[col].fillna(median_value)

display(df_ROMA.isnull().sum())
```

```python
# Calculate the mode for 'has_availability'
mode_has_availability = df_ROMA['has_availability'].mode()[0]

# Fill missing values with the calculated mode
df_ROMA['has_availability'] = df_ROMA['has_availability'].fillna(mode_has_availability)

display(df_ROMA.isnull().sum())
```

```python
# Calculate the median for 'estimated_revenue_l365d'
median_estimated_revenue = df_ROMA['estimated_revenue_l365d'].median()

# Fill missing values with the calculated median
df_ROMA['estimated_revenue_l365d'] = df_ROMA['estimated_revenue_l365d'].fillna(median_estimated_revenue)

# Establecer un código de fecha para first y last_review
df_ROMA["first_review"] = df_ROMA["first_review"].fillna("00/00/00")
df_ROMA["last_review"] = df_ROMA["last_review"].fillna("00/00/00")

#Establecer con 0 todos los values que no cuentan con un review dentro este tipo de variables el dataframe
df_ROMA['reviews_per_month'] = df_ROMA['reviews_per_month'].fillna(0)
df_ROMA['review_scores_rating'] = df_ROMA['review_scores_rating'].fillna(0)
df_ROMA['review_scores_accuracy'] = df_ROMA['review_scores_accuracy'].fillna(0)
df_ROMA['review_scores_cleanliness'] = df_ROMA['review_scores_cleanliness'].fillna(0)
df_ROMA['review_scores_checkin'] = df_ROMA['review_scores_checkin'].fillna(0)
df_ROMA['review_scores_communication'] = df_ROMA['review_scores_communication'].fillna(0)
df_ROMA['review_scores_location'] = df_ROMA['review_scores_location'].fillna(0)
df_ROMA['review_scores_value'] = df_ROMA['review_scores_value'].fillna(0)

#Establecer una valor "UNKNOWN" a todas los valores nulos dentro de la columna 'license'
df_ROMA["license"] = df_ROMA["license"].fillna("UNKNOWN")

display(df_ROMA.isnull().sum())
```

**PROCEDIMIENTO "IQR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=df_quantitative

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Obtenemos datos limpios del Dataframe
df_cuartiles = df_quantitative[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
```

```python
# Identify integer columns in df_cuartiles
integer_columns = df_cuartiles.select_dtypes(include=np.number).columns

# Impute outliers (NaN values) in integer columns with the median
for col in integer_columns:
    median_value = df_cuartiles[col].median()
    df_cuartiles[col] = df_cuartiles[col].fillna(median_value)

# Display the null value counts to verify imputation
display(df_cuartiles.isnull().sum())

# Assign the cleaned dataframe to a new variable name
df_quantitative_filled_outliers = df_cuartiles
```

```python
#Corroboramos valores nulos (Atípicos) del dataframe FINAL
display(df_quantitative_filled_outliers.isnull().sum())
```

```python
def get_top_correlated_pairs(correlation_matrix, n=10, exclude_above=0.90):
    """
    Extracts the top N most correlated pairs from a correlation matrix,
    excluding self-correlations and pairs above a certain absolute correlation.
    """
    unstacked = correlation_matrix.unstack()
    sorted_unstacked = unstacked.sort_values(kind="quicksort", ascending=False)

    # Filter out correlations above the exclusion threshold
    filtered_sorted_unstacked = sorted_unstacked[abs(sorted_unstacked) <= exclude_above]

    # Drop self-correlations and duplicate pairs
    unique_pairs = filtered_sorted_unstacked[filtered_sorted_unstacked.index.get_level_values(0) != filtered_sorted_unstacked.index.get_level_values(1)]
    unique_pairs = unique_pairs.drop_duplicates()

    # Get the top N pairs
    top_pairs = unique_pairs.head(n)

    return top_pairs

# Get top 10 correlated pairs for each room type
top_10_entire_home = get_top_correlated_pairs(correlation_matrix_entire_home)
top_10_hotel_room = get_top_correlated_pairs(correlation_matrix_hotel_room)
top_10_private_room = get_top_correlated_pairs(correlation_matrix_private_room)
top_10_shared_room = get_top_correlated_pairs(correlation_matrix_shared_room)

# Display the top 10 correlated pairs for each room type
print("Top 10 Correlated Pairs for 'Entire home/apt':")
display(top_10_entire_home)

print("\n" + "="*50 + "\n") # Separator

print("Top 10 Correlated Pairs for 'Hotel room':")
display(top_10_hotel_room)

print("\n" + "="*50 + "\n") # Separator

print("Top 10 Correlated Pairs for 'Private room':")
display(top_10_private_room)

print("\n" + "="*50 + "\n") # Separator

print("Top 10 Correlated Pairs for 'Shared room':")
display(top_10_shared_room)
```

```python
# List of dependent variables
dependent_variables = [
    'review_scores_rating',
    'host_acceptance_rate',
    'host_is_superhost',
    'host_total_listings_count',
    'accommodates',
    'bedrooms',
    'price',
    'review_scores_value',
    'bathrooms',
    'reviews_per_month'
]

# Select only numeric columns from the merged dataframe
numeric_df_merged = df_merged.select_dtypes(include=np.number)

# Calculate the correlation matrix for the numeric columns
correlation_matrix_merged = numeric_df_merged.corr()

# Create a dictionary to store the top correlations for each dependent variable
top_correlations = {}

# Iterate through the dependent variables
for dependent_var in dependent_variables:
    if dependent_var in correlation_matrix_merged.columns:
        # Get the correlations for the dependent variable, excluding self-correlation
        correlations = correlation_matrix_merged[dependent_var].drop(dependent_var)

        # Filter out correlations greater than 95% (absolute value)
        filtered_correlations = correlations[abs(correlations) <= 0.90]

        # Sort the correlations by absolute value in descending order
        sorted_correlations = filtered_correlations.abs().sort_values(ascending=False)

        # Get the top 5 correlations
        top_5_correlations = sorted_correlations.head(5)

        # Store the results
        top_correlations[dependent_var] = top_5_correlations
    else:
        print(f"Warning: '{dependent_var}' is not a numeric column in the dataframe.")


# Display the top 5 correlations for each dependent variable in a table format
print("Top 5 Highest Correlations for Specified Dependent Variables (excluding > 95% correlation):")
for dependent_var, correlations in top_correlations.items():
    print(f"\nDependent Variable: {dependent_var}")
    display(correlations.to_frame(name='Correlation Coefficient'))
```

**REGRESIONES MÚLTIPLES**

1. Seleccionar variables dependientes e independientes

```python
selected_independent_variables = {}

for dependent_var in dependent_variables:
    if dependent_var in top_correlations:
        sorted_correlations = top_correlations[dependent_var]
        selected_independent_variables[dependent_var] = []

        current_selected = []
        for indep_var, correlation_value in sorted_correlations.items():
            is_multicollinear = False
            for selected in current_selected:
                # Check for multicollinearity (e.g., correlation > 0.8)
                if abs(correlation_matrix_merged.loc[indep_var, selected]) > 0.8:
                    is_multicollinear = True
                    break

            if not is_multicollinear and len(current_selected) < 3:
                current_selected.append(indep_var)

            if len(current_selected) == 3:
                break

        selected_independent_variables[dependent_var] = current_selected

# Print the selected independent variables for verification
print("Selected Independent Variables for Regression Models:")
for dependent_var, indep_vars in selected_independent_variables.items():
    print(f"{dependent_var}: {indep_vars}")
```

```python
import statsmodels.api as sm

fitted_models = {}

for dependent_var, independent_vars in selected_independent_variables.items():
    if independent_vars: # Only build a model if there are independent variables selected
        y = df_merged[dependent_var]
        X = df_merged[independent_vars]
        X = sm.add_constant(X) # Add a constant (intercept) to the independent variables

        model = sm.OLS(y, X)
        fitted_model = model.fit()
        fitted_models[dependent_var] = fitted_model

# You can optionally print summaries of the fitted models to verify
# for dependent_var, model in fitted_models.items():
#     print(f"\nRegression Model Summary for Dependent Variable: {dependent_var}")
#     print(model.summary())
```

2. Regresiones por variable Independiente (Coeficiente de determinación R2)

```python
regression_results = []

for dependent_var, model in fitted_models.items():
    r_squared = model.rsquared

    # Construct the regression formula string
    formula = f"{dependent_var} = {model.params['const']:.4f}"
    for indep_var, coef in model.params.drop('const').items():
        formula += f" + {coef:.4f} * {indep_var}"

    regression_results.append({
        'Dependent Variable': dependent_var,
        'Independent Variables': selected_independent_variables[dependent_var],
        'Regression Formula': formula,
        'R-squared': r_squared
    })

# Display the results in a table
display(pd.DataFrame(regression_results))
```

```python
from scipy.stats import pearsonr

correlation_results_r = []

for dependent_var, model in fitted_models.items():
    # Get the actual values
    y_true = df_merged[dependent_var]
    # Get the predicted values
    y_pred = model.predict(sm.add_constant(df_merged[selected_independent_variables[dependent_var]]))

    # Calculate the Pearson correlation coefficient and p-value
    correlation_r, p_value = pearsonr(y_true, y_pred)

    # Construct the regression formula string
    formula = f"{dependent_var} = {model.params['const']:.4f}"
    for indep_var, coef in model.params.drop('const').items():
        formula += f" + {coef:.4f} * {indep_var}"

    correlation_results_r.append({
        'Dependent Variable': dependent_var,
        'Independent Variables': selected_independent_variables[dependent_var],
        'Regression Formula': formula,
        'Pearson R': correlation_r
    })

# Display the results in a table
display(pd.DataFrame(correlation_results_r))
```
